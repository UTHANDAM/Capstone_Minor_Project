# Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import skew, kurtosis, spearmanr, pearsonr
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

# DATA LOADING AND PREPARATION

```python
# Load the datasets (skip comment lines and header row)
female = np.genfromtxt('nhanes_adult_female_bmx_2020.csv', delimiter=',', skip_header=19)
male = np.genfromtxt('nhanes_adult_male_bmx_2020.csv', delimiter=',', skip_header=19)
```

```python
print(f"\nFemale dataset shape: {female.shape}")
print(f"Male dataset shape: {male.shape}")
```

```python
print("Column Information:")
print("\n1. Weight (kg)")
print("2. Standing Height (cm)")
print("3. Upper Arm Length (cm)")
print("4. Upper Leg Length (cm)")
print("5. Arm Circumference (cm)")
print("6. Hip Circumference (cm)")
print("7. Waist Circumference (cm)")
```

```python
print(f"Total female participants: {female.shape[0]}")
print(f"Total male participants: {male.shape[0]}")
```

# WEIGHT DISTRIBUTION HISTOGRAMS

```python
# Extract weights
female_weights = female[:, 0]
male_weights = male[:, 0]

# Determine common x-axis limits for comparison
xlim = (30, 210)
```

```python
# Create histograms with two subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# Female weights histogram (top subfigure)
axes[0].hist(female_weights, bins=50, color='#FF6B9D', alpha=0.7, edgecolor='black')
axes[0].set_xlim(xlim)
axes[0].set_xlabel('Weight (kg)', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].set_title('Distribution of Female Weights', fontsize=13, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].axvline(female_weights.mean(), color='red', linestyle='--', linewidth=2, 
                label=f'Mean: {female_weights.mean():.2f} kg')
axes[0].legend()

# Male weights histogram (bottom subfigure)
axes[1].hist(male_weights, bins=50, color='#4A90E2', alpha=0.7, edgecolor='black')
axes[1].set_xlim(xlim)
axes[1].set_xlabel('Weight (kg)', fontsize=11)
axes[1].set_ylabel('Frequency', fontsize=11)
axes[1].set_title('\n Distribution of Male Weights', fontsize=13, fontweight='bold')
axes[1].grid(True, alpha=0.3)
axes[1].axvline(male_weights.mean(), color='red', linestyle='--', linewidth=2, 
                label=f'Mean: {male_weights.mean():.2f} kg')
axes[1].legend()
plt.tight_layout()
plt.savefig('weight_histograms.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print(f"Female weight range: {female_weights.min():.2f} - {female_weights.max():.2f} kg")
print(f"Male weight range: {male_weights.min():.2f} - {male_weights.max():.2f} kg")
```

# COMPARATIVE BOX-AND-WHISKER PLOT

```python
fig, ax = plt.subplots(figsize=(10, 7))
box_data = [female_weights, male_weights]
bp = ax.boxplot(box_data, labels=['Female', 'Male'], patch_artist=True,
                widths=0.6, showmeans=True,
                medianprops=dict(color='red', linewidth=2),
                meanprops=dict(marker='D', markerfacecolor='green', markersize=8))
colors = ['#FF6B9D', '#4A90E2']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel('Weight (kg)', fontsize=12)
ax.set_title('Comparison of Male and Female Weight Distributions', 
             fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.savefig('weight_boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Calculate quartiles for interpretation
female_q1, female_median, female_q3 = np.percentile(female_weights, [25, 50, 75])
male_q1, male_median, male_q3 = np.percentile(male_weights, [25, 50, 75])
```

```python
print("DISCUSSION:")
print("\nThe box-and-whisker plot reveals clear sex differences in weight distribution.")
print(f"Males exhibit consistently higher weights across all quartiles (median: {male_median:.2f} kg),compared to females (median: {female_median:.2f} kg), a difference of {male_median - female_median:.2f} kg.")
print(f"Males also show greater variability (IQR: {male_q3 - male_q1:.2f} kg vs {female_q3 - female_q1:.2f} kg),reflecting more diverse body compositions. Both distributions show outliers on the     higher end,indicating individuals with extreme obesity.")
```

#  NUMERICAL AGGREGATES AND DISTRIBUTION ANALYSIS

```python
# Measures of Location
female_mean = np.mean(female_weights)
female_median = np.median(female_weights)
male_mean = np.mean(male_weights)
male_median = np.median(male_weights)
```

```python
# Measures of Dispersion
female_var = np.var(female_weights, ddof=1)
female_std = np.std(female_weights, ddof=1)
female_cv = (female_std / female_mean) * 100
male_var = np.var(male_weights, ddof=1)
male_std = np.std(male_weights, ddof=1)
male_cv = (male_std / male_mean) * 100
```

```python
# Measures of Shape
female_skewness = skew(female_weights)
female_kurtosis = kurtosis(female_weights, fisher=True)  # Excess kurtosis
male_skewness = skew(male_weights)
male_kurtosis = kurtosis(male_weights, fisher=True)
```

```python
print("\n FEMALE WEIGHTS: ")
print(f"Location: Mean={female_mean:.2f} kg, Median={female_median:.2f} kg")
print(f"Dispersion: Variance={female_var:.2f}, Std Dev={female_std:.2f} kg, CV={female_cv:.2f}%")
print(f"Shape: Skewness={female_skewness:.4f}, Excess Kurtosis={female_kurtosis:.4f}")
```

```python
print("\n MALE WEIGHTS:")
print(f"Location: Mean={male_mean:.2f} kg, Median={male_median:.2f} kg")
print(f"Dispersion: Variance={male_var:.2f}, Std Dev={male_std:.2f} kg, CV={male_cv:.2f}%")
print(f"Shape: Skewness={male_skewness:.4f}, Excess Kurtosis={male_kurtosis:.4f}")
```

```python
print("\nDISCUSSION:")
print(f"Both distributions exhibit POSITIVE SKEWNESS (Female: {female_skewness:.3f}, Male: {male_skewness:.3f}),indicating right-tailed distributions with more lighter individuals and some heavier outliers.")
print(f"Both show LEPTOKURTIC distributions (excess kurtosis > 0), meaning heavier tails and more extreme values than a normal distribution. Males are heavier on average ({male_mean:.2f} kg vs{female_mean:.2f} kg), but surprisingly show slightly less variability (CV: {male_cv:.2f}% vs {female_cv:.2f}%),suggesting more consistent body mass among males relative to their mean.")
```

# ADD BMI COLUMN TO FEMALE MATRIX

```python
# Calculate BMI = weight (kg) / (height (m))^2
female_weight = female[:, 0]
female_height_m = female[:, 1] / 100 
female_bmi = female_weight / (female_height_m ** 2)
female = np.column_stack((female, female_bmi))
print(f"Female dataset new shape: {female.shape}")
print(f"\nBMI Statistics: Mean={female_bmi.mean():.2f}, Median={np.median(female_bmi):.2f}")
print(f"Range: {female_bmi.min():.2f} - {female_bmi.max():.2f}")
```

```python
underweight = np.sum(female_bmi < 18.5)
normal = np.sum((female_bmi >= 18.5) & (female_bmi < 25))
overweight = np.sum((female_bmi >= 25) & (female_bmi < 30))
obese = np.sum(female_bmi >= 30)
```

```python
print(f"\nBMI Categories: Underweight={underweight} ({underweight/len(female_bmi)*100:.1f}%), "
      f"Normal={normal} ({normal/len(female_bmi)*100:.1f}%), "
      f"Overweight={overweight} ({overweight/len(female_bmi)*100:.1f}%), "
      f"Obese={obese} ({obese/len(female_bmi)*100:.1f}%)")
```

```python
print("\nDISCUSSION:")
print(f"The BMI analysis reveals concerning obesity prevalence: {obese/len(female_bmi)*100:.1f}% of females")
print(f"are obese (BMI  30), and only {normal/len(female_bmi)*100:.1f}% have normal BMI. This reflects")
print("the widespread obesity epidemic in the United States adult female population.")
```

# CREATE STANDARDIZED DATASET (Z-SCORES)

```python
# Calculate z-scores for all columns
zfemale = np.zeros_like(female)
for col in range(female.shape[1]):
    col_mean = np.mean(female[:, col])
    col_std = np.std(female[:, col], ddof=1)
    zfemale[:, col] = (female[:, col] - col_mean) / col_std
print(f"\nStandardized dataset (zfemale) shape: {zfemale.shape}")
print("\nVerification (mean  0, std  1):")
print(f"Weight: Mean={zfemale[:, 0].mean():.6f}, Std={zfemale[:, 0].std(ddof=1):.6f}")
print(f"BMI: Mean={zfemale[:, 7].mean():.6f}, Std={zfemale[:, 7].std(ddof=1):.6f}")
```

```python
print("\n**DISCUSSION:**")
print("Z-score standardization transforms all variables to have mean=0 and std=1, allowing")
print("direct comparison of measurements on different scales. Positive z-scores indicate values")
print("above the mean, negative below. This is essential for correlation analysis and identifying outliers.")
```

# SCATTERPLOT MATRIX AND CORRELATIONS

```python
# Select columns: Height(1), Weight(0), Waist(6), Hip(5), BMI(7) from zfemale
selected_indices = [1, 0, 6, 5, 7]  # Height, Weight, Waist, Hip, BMI
selected_names = ['Height', 'Weight', 'Waist', 'Hip', 'BMI']
selected_data = zfemale[:, selected_indices]
n_vars = len(selected_indices)
fig, axes = plt.subplots(n_vars, n_vars, figsize=(16, 16))
for i in range(n_vars):
    for j in range(n_vars):
        ax = axes[i, j]
        if i == j:
            # Diagonal: histograms
            ax.hist(selected_data[:, i], bins=40, color='steelblue', alpha=0.7, edgecolor='black')
            ax.set_ylabel('Frequency', fontsize=8)
        else:
            # Off-diagonal: scatter plots
            ax.scatter(selected_data[:, j], selected_data[:, i], alpha=0.3, s=1, color='darkblue')
            ax.grid(True, alpha=0.3)
        if i == n_vars - 1:
            ax.set_xlabel(selected_names[j], fontsize=10, fontweight='bold')
        if j == 0:
            ax.set_ylabel(selected_names[i], fontsize=10, fontweight='bold')
        if i < n_vars - 1:
            ax.set_xticklabels([])
        if j > 0:
            ax.set_yticklabels([])
plt.suptitle('Scatterplot Matrix for Female Body Measurements (Standardized)', 
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.savefig('scatterplot_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Calculate Pearson correlation coefficients
pearson_corr = np.corrcoef(selected_data.T)
print("\n--- PEARSON CORRELATION MATRIX ---")
print(f"{'':10}", end='')
for name in selected_names:
    print(f"{name:>10}", end='')
print()
for i, name in enumerate(selected_names):
    print(f"{name:10}", end='')
    for j in range(n_vars):
        print(f"{pearson_corr[i, j]:10.4f}", end='')
    print()
```

```python
# Calculate Spearman correlation coefficients
spearman_corr = np.zeros((n_vars, n_vars))
for i in range(n_vars):
    for j in range(n_vars):
        spearman_corr[i, j], _ = spearmanr(selected_data[:, i], selected_data[:, j])

print("\n--- SPEARMAN CORRELATION MATRIX ---")
print(f"{'':10}", end='')
for name in selected_names:
    print(f"{name:>10}", end='')
print()
for i, name in enumerate(selected_names):
    print(f"{name:10}", end='')
    for j in range(n_vars):
        print(f"{spearman_corr[i, j]:10.4f}", end='')
    print()
```

```python
print("\n**DISCUSSION:**")
print(f"Strong correlations exist between Weight-BMI (r={pearson_corr[1, 4]:.3f}), Waist-BMI (r={pearson_corr[2, 4]:.3f}),")
print(f"and Hip-BMI (r={pearson_corr[3, 4]:.3f}), confirming that BMI reflects overall adiposity. The Waist-Hip")
print(f"correlation (r={pearson_corr[2, 3]:.3f}) shows both increase together with fat accumulation. Height shows")
print(f"weak correlation with BMI (r={pearson_corr[0, 4]:.3f}), confirming BMI properly normalizes for height.")
print("Pearson and Spearman correlations are nearly identical, indicating predominantly linear relationships.")
```

# ADD WAIST-TO-HEIGHT AND WAIST-TO-HIP RATIOS

```python
# For FEMALE data
female_waist = female[:, 6]
female_height = female[:, 1]
female_whr_height = female_waist / female_height
female_hip = female[:, 5]
female_whr_hip = female_waist / female_hip
female = np.column_stack((female, female_whr_height, female_whr_hip))
```

```python
# For MALE data
male_waist = male[:, 6]
male_height = male[:, 1]
male_whr_height = male_waist / male_height
male_hip = male[:, 5]
male_whr_hip = male_waist / male_hip
male = np.column_stack((male, male_whr_height, male_whr_hip))
```

```python
print(f"\nFemale WHtR: Mean={female_whr_height.mean():.4f}, Median={np.median(female_whr_height):.4f}")
print(f"Female WHR: Mean={female_whr_hip.mean():.4f}, Median={np.median(female_whr_hip):.4f}")
print(f"Male WHtR: Mean={male_whr_height.mean():.4f}, Median={np.median(male_whr_height):.4f}")
print(f"Male WHR: Mean={male_whr_hip.mean():.4f}, Median={np.median(male_whr_hip):.4f}")
```

```python
# Health risk assessment
female_high_whr = np.sum(female_whr_hip > 0.85)
male_high_whr = np.sum(male_whr_hip > 0.90)
female_high_wthr = np.sum(female_whr_height > 0.5)
male_high_wthr = np.sum(male_whr_height > 0.5)
```

```python
print(f"\nHealth Risk Assessment:")
print(f"Females with WHR > 0.85 (high risk): {female_high_whr/len(female_whr_hip)*100:.1f}%")
print(f"Males with WHR > 0.90 (high risk): {male_high_whr/len(male_whr_hip)*100:.1f}%")
print(f"Females with WHtR > 0.5 (increased risk): {female_high_wthr/len(female_whr_height)*100:.1f}%")
print(f"Males with WHtR > 0.5 (increased risk): {male_high_wthr/len(male_whr_height)*100:.1f}%")
```

```python
print("\n**DISCUSSION:**")
print(f"Males have higher WHR ({male_whr_hip.mean():.3f} vs {female_whr_hip.mean():.3f}), reflecting central")
print("(abdominal) fat distribution ('apple shape'), while females show lower WHR ('pear shape').")
print("This sex difference is hormonally driven and explains higher cardiovascular risk in males.")
print(f"Over 80% of both groups exceed risk thresholds, indicating widespread obesity.")
```

# COMPARATIVE BOX-AND-WHISKER PLOT FOR RATIOS

```python
box_data = [female_whr_height, male_whr_height, female_whr_hip, male_whr_hip]
labels = ['Female\nWHtR', 'Male\nWHtR', 'Female\nWHR', 'Male\nWHR']

fig, ax = plt.subplots(figsize=(12, 8))
bp = ax.boxplot(box_data, labels=labels, patch_artist=True,
                widths=0.5, showmeans=True,
                medianprops=dict(color='red', linewidth=2.5),
                meanprops=dict(marker='D', markerfacecolor='green', markersize=10))

colors = ['#FF6B9D', '#4A90E2', '#FF6B9D', '#4A90E2']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)
    # Add reference lines for risk thresholds
ax.axhline(y=0.5, color='purple', linestyle='--', linewidth=2, alpha=0.7, 
           label='WHtR Risk Threshold (0.5)')
ax.axhline(y=0.85, color='orange', linestyle='--', linewidth=2, alpha=0.7, 
           label='WHR Risk (Female: 0.85)')
ax.axhline(y=0.90, color='brown', linestyle='--', linewidth=2, alpha=0.7, 
           label='WHR Risk (Male: 0.90)')

ax.set_ylabel('Ratio Value', fontsize=13, fontweight='bold')
ax.set_title('Comparison of Waist-to-Height and Waist-to-Hip Ratios by Sex', 
             fontsize=15, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
ax.legend(loc='upper left', fontsize=10)
plt.tight_layout()
plt.savefig('ratio_boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("\n**DISCUSSION:**")
print("The boxplot reveals distinct patterns: (1) Females have HIGHER WHtR than males, indicating")
print("relatively larger waist circumference for their height; (2) Males have HIGHER WHR, showing")
print("greater waist-to-hip ratio (central obesity); (3) Both WHtR distributions exceed the 0.5 threshold")
print("(majority above risk level); (4) Males exceed WHR threshold (0.9) more than females exceed theirs (0.85).")
print("These patterns reflect sex-specific fat distribution: males accumulate abdominal fat (higher")
print("cardiovascular risk), while females accumulate hip/thigh fat (lower risk despite higher WHtR).")
```

# ADVANTAGES AND DISADVANTAGES ANALYSIS

```python
print("\n**BMI (BODY MASS INDEX)**")
print("\nAdvantages:")
print("   Simple calculation requiring only weight and height")
print("   Universally accepted with extensive reference data")
print("   Cost-effective for population screening")
print("   Correlates with health outcomes at population level")
print("\nDisadvantages:")
print("   Cannot distinguish muscle mass from fat mass")
print("   Ignores fat distribution (visceral vs subcutaneous)")
print("   Less accurate for athletes, elderly, and certain ethnic groups")
print("   May misclassify 'normal weight obese' individuals")
```

```python
print("\n**WAIST-TO-HEIGHT RATIO (WHtR)**")
print("\nAdvantages:")
print("   Better predictor of cardiovascular risk than BMI")
print("   Simple universal threshold (0.5) across all demographics")
print("   Accounts for height variation")
print("   Identifies central obesity early")
```

```python
print("\n**WAIST-TO-HEIGHT RATIO (WHtR)**")
print("\nAdvantages:")
print("   Better predictor of cardiovascular risk than BMI")
print("   Simple universal threshold (0.5) across all demographics")
print("   Accounts for height variation")
print("   Identifies central obesity early")
print("\nDisadvantages:")
print("   Measurement technique sensitive (where to measure waist)")
print("   Cannot distinguish fat from muscle")
print("   Less historical data than BMI")
print("   Self-measurement can be inaccurate")
```

```python
print("\n**WAIST-TO-HIP RATIO (WHR)**")
print("\nAdvantages:")
print("   Assesses fat distribution patterns (apple vs pear shape)")
print("   Better mortality predictor than BMI in many studies")
print("   Captures sex-specific fat distribution")
print("   Strong association with metabolic syndrome")

print("\nDisadvantages:")
print("   Requires two measurements (doubles error potential)")
print("   Different thresholds for sex and ethnicity")
print("   MAJOR LIMITATION: Insensitive to weight changes (both waist and hip")
print("    decrease proportionally, ratio may stay same)")
print("   Hip measurement less standardized, more difficult")
```
